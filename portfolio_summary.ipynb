
# Data Analysis Portfolio Summary
This notebook summarizes all five projects completed in the portfolio, including objectives, datasets, key insights, and recommendations.



## Project 1: Business Sales Analysis
**Objective:** Analyze sales trends, customer behavior, and product performance.

**Dataset:** Simulated retail sales data

**Analysis Type:** Exploratory + Customer & Product Analytics

**Key Insights:**
- Electronics category contributes the highest revenue
- Repeat customers generate significant sales
- Laptop is the top-selling product by value

**Recommendations:**
- Focus marketing on high-value electronics products
- Introduce loyalty programs for repeat customers
- Bundle accessories with laptops to increase cart value
