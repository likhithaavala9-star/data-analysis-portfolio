```python
customer_sales = df.groupby('Customer_ID')['Total_Sales'].sum().sort_values(ascending=False)

customer_sales.plot(kind='bar', title="Customer-wise Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
category_sales = df.groupby('Category')['Total_Sales'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%', title="Category Sales Share")
plt.ylabel("")
plt.show()
```


The Electronics category dominates total sales contribution,
while Fashion products account for a smaller share of overall revenue.
This highlights category-level revenue imbalance.
